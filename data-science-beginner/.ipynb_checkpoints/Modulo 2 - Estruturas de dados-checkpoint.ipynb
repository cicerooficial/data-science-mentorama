##### Slicing em listas

```python
lista_nomes
```

```python
listas_nomes[:2]
```

##### Funções nativas para listas

```python
# Já vimos que podemos adicionar elementos na lista concateando com outra

linguagens = ["Python","SQL","R"]

linguagens_1 = linguagens + ["Java","C++"]

linguagens_1
```

```python
linguagens_1.clear()

print(linguagens_1)
```

```python
# listas.+ tab mostra funções nativas para listas - aprendemos mais sobre funções no fim do mósulo
# append adiciona elemento no final da lista

linguagens.append("Java")
linguagens.append("C++")

linguagens
```

```python
linguagens.append(["Java","C++"])
linguagens
```

```python
# extend adiciona outra lista com elementos da lista original e não como uma lista 

linguagens.extend(["Java","C++"])
linguagens
```

```python
# insert adiciona elementos em posição específicas da lista

linguagens.insert(1,"Scala")

linguagens
```

```python
# remove Deleta o primeiro elemnto encontrado na lista com esse nome

linguagens.remove("C++")

linguagens
```

```python
# pop remove elementos da lista usando indice

linguagens.pop(len(linguagens)-1)
```

```python
linguagens
```

```python
linguagens .pop(2)

print(linguagens)
```

```python
# count conta a quantidade de elementos da lista

linguagens.count("Java")
```

```python
# sort ordena os elementos da lista (Se for String, faz por ordem alfabética)

lista_nomes.sort()
lista_nomes
```

```python
lista_range.sort()
lista_range
```
